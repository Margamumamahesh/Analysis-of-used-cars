```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df=pd.read_csv("Used_cars.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['CITY']=df['CITY'].ffill()
```

```python
df["PRICE"] = df["PRICE"].str.split("₹").str[0]
```

```python
df.head()
```

```python
df
```

```python
df["PRICE"] = df["PRICE"].replace(",", "", regex=True).astype(int)
```

```python
df["MONTHLY_EMI"] = df["MONTHLY_EMI"].replace(",", "", regex=True).astype(int)
```

```python
df["KMs_DRIVEN"] = df["KMs_DRIVEN"].replace(",", "", regex=True).astype(int)
```

```python
df.to_csv("Used_cars_India.csv",index=False)
```
