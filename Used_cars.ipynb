```python
brand=['maruti','honda','hyundai','tata', 'renault']
city=['1692','3686','2378','2','2423']
```

```python
year=[]
Brand=[]
Model = []
kms=[]
fuel=[]
ownership=[]
Spec = []
Price = []
City = []
Emi = []

for i in city:
    for j in brand:
        url = f"https://www.cars24.com/buy-used-car?f=make%3A%3D%3A{j}&sort=bestmatch&serveWarrantyCount=true&gaId=18592817.1696917583&listingSource=TabFilter&storeCityId={i}"
        page = requests.get(url)
        soup = BeautifulSoup(page.text, 'html.parser')
        
        for t in soup.find_all('h2',{'class':"_2lmIw"}):
            txt=t.text
            
            yr=re.findall("(\d+)\s",txt)
            if yr:
                year.append(yr[0])
            else:
                year.append(np.nan)
                
            br=re.findall("\d+\s(\w+)\s",txt)
            if br:
                Brand.append(br[0])
            else:
                Brand.append(np.nan)
            
            md=re.findall("['Maruti'|'Honda'|'Hyundai'|'Tata'| 'Renault']\s(.+)",txt)
            if md:
                Model.append(md[0])
            else:
                Model.append(np.nan)
                
        for t in soup.find_all('ul',{'class':"_13yb6"}):
            txt=t.text
            
            km=re.findall("(.+)\skm",txt)
            if km:
                kms.append(km[0])
            else:
                kms.append(np.nan)
                
            fu=re.findall("(Diesel|Petrol|CNG)",txt)
            if fu:
                fuel.append(fu[0])
            else:
                fuel.append(np.nan)
            
            os=re.findall("km(.+)\sOwner",txt)
            if md:
                ownership.append(os[0])
            else:
                ownership.append(np.nan)
            
        for t in soup.find_all('div',{'class':"_18ToE"}):
            txt=t.text
            
            pr=re.findall("₹(.+)",txt)
            if pr:
                Price.append(pr[0])
            else:
                Price.append(np.nan)
                
        for t in soup.find_all('span',{'class':"_3DYbK"}):
            txt=t.text
            
            ct=re.findall("(Hyderabad|Ahmedabad|New Delhi|Mumbai|Pune)",txt)
            if ct:
                City.append(ct[0])
            else:
                City.append(np.nan)
                
        for t in soup.find_all('div',{'class':"_1Em-A"}):
            txt=t.text
            
            em=re.findall("₹(.+)/month",txt)
            if pr:
                Emi.append(em[0])
            else:
                Emi.append(np.nan)
                
```

```python
dic={'YEAR':year,
'BRAND':Brand,
'MODEL':Model,
'KMs_DRIVEN':kms,
'FUEL_TYPE':fuel,
'OWNER':ownership,
'PRICE':Price,
'CITY':City,
'MONTHLY_EMI':Emi}
```

```python
df=pd.DataFrame(dic)
```
